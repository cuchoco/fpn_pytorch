```python
with open('./data/annotation.json', 'r') as file:
    ann = json.load(file)

ds = BrainDataset(ann)
```

```python
import matplotlib.pyplot as plt
import cv2

def draw_box(img, bboxs, labels=None):
    '''bbox form (x_min, y_min, x_max, y_max)'''
    if labels:
        for box, label in zip(bboxs, labels):
            x_min,y_min,x_max,y_max = box
            cv2.rectangle(img,(x_min,y_min),(x_max, y_max),(0,0,255),2)
            cv2.putText(img, f'{label}', (x_min, y_min-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,0,255), 2)
    
    else:
        for box in bboxs:
            x_min,y_min,x_max,y_max = box
            cv2.rectangle(img,(x_min,y_min),(x_max, y_max),(0,0,255),2)
    return img

idx = 120
img = ds[idx]['img']
bbox = ds[idx]['annot']
bbox = bbox[:,:4]


plt.imshow(draw_box(img, bbox))
```
